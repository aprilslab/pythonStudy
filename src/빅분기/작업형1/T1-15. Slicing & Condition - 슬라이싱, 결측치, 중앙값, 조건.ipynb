f1의 결측치를 중앙값으로 채운다.\

```python
import pandas as pd

df = pd.read_csv('../dataset/basic1.csv')
df.head()
```

## Solution

```python
df.isnull().sum()
```

```python
df2 = df.sort_values('age', ascending=False).reset_index(drop=True)
df2 = df2[:20]

# df2 = df.sort_values('age', ascending=False).head(20)
df2
```

```python
df2['f1'] = df2['f1'].fillna(df2['f1'].median())
df2
```

```python
df2[(df2['f4'] == 'ISFJ') & (df2['f5'] >= 20)]['f1'].mean()
```

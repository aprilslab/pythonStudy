IQR을 이상치를 정제하기위한 판단 기준으로 사용하기도 한다.

## 상자그림 그리기

**(1) iris변수들**

```python
# 데이터 가져오기
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris['class'] = load_iris().target
iris['class'] = iris['class'].map({0:'Setosa', 1:'Versicolour', 2:'Virginica'})

# boxplot 그리기
plt.boxplot(iris.drop(columns='class'))
plt.show()
```

**(2) 하나의 컬럼만 표현**

`whis`는 IQR에 곲할수를 지정하는 인수 default=1.5

```python
plt.boxplot(iris['sepal width (cm)'], whis=1.5)
plt.title('sepal width (cm)')
plt.show()
```

**(3) 한 컬럼을 카테고리 변수에 따라 분리**

'class' 컬럼의 카테고리에 따른 'sepal width (cm)'값의 차이

```python
iris[['sepal width (cm)', 'class']].boxplot(by='class')
plt.show()
```

**(4) Out[3]의 그래프를 seaborn 라이브러리로 더욱 시각적으로 꾸밈**

```python
import seaborn as sns
sns.boxplot(x='class', y='sepal width (cm)', data=iris)
plt.show()
```

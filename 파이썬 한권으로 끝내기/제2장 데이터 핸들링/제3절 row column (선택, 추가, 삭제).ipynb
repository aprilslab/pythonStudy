## row/colum 선택

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris[1:4]
```

```python
iris['sepal length (cm)'].head(4)
```

```python
iris[['sepal length (cm)', 'sepal width (cm)']].head(4)
```

- DataFrame['컬럼명']: 하나의 컬럼을 Series 형식으로 출력
- DataFrame[['컬럼명1', '컬럼명2']]: 여러개의 컬럼을 DataFrame 형식으로 출력

### iloc

정수로 된 위치 기반 인덱스(int location)와 행과 열을 선택하고자 할 때는 iloc 속성을 사용한다.\
인수는 하나 혹은 두 개가 사용되며 `정수, 리스트, 슬라이싱, bool array`가 들어갈 수 있다.

- 인수 하나: `행`지정
- 인수 두 개: 순서대로 `행`과 `열`

```python
iris.iloc[1:4]
```

```python
iris.iloc[[1,3,5], 2:4]
```

```python
iris.iloc[:, [True, True, False, True]]
```

### loc

레이블(인덱스 이름 혹은 컬럼명)로 행과 열을 선택하고자 할 때는 loc[] 속성을 사용한다.\
파이썬 기본 슬라이싱 방법과 달리 end를 포함한다.

```python
iris.loc[1:3]
```

```python
iris.loc[[1,2], 'sepal length (cm)': 'petal length (cm)']
```

## row/column 추가

### 선택한 값 변경하기

```python
score = pd.DataFrame({
    '국어': [100, 80],
    '수학': [75, 90],
    '영어': [90, 95]
}, index=['장화', '홍련'])

score.loc['홍련', '영어'] = 100
score['국어'] = score['국어'] - 5
score
```

### row 추가

append() 사용\
인수: DataFrame, Series, Dictionary\
Series를 사용하는 경우 Series의 name이 추가되는 행의 index가 된다.

```python
new_students = pd.DataFrame({
    '국어': [70, 85],
    '수학': [65, 100],
    '영어': [95, 65]
}, index=['콩쥐', '팥쥐'])
score = score.append(new_students)
new_student1 = pd.Series({
    '국어': 85,
    '수학': 55,
    '영어': 95
}, name='해님')
score = score.append(new_student1)
score
```

```python
new_student2 = {
    '국어': 75,
    '수학': 80,
    '영어': 80
}
score.append(new_student2, ignore_index=True)
```

### column 추가

```python
science = [80, 70, 90, 85, 75]
score['과학'] = science
score['학년'] = 1
score
```

```python
score['과학'] = score['과학'] + 5
score['총점'] = score['국어'] + score['수학'] + score['영어'] + score['과학']
score
```

## row/column 삭제

```python
score.drop('장화', inplace=True)
score.drop(columns = ['과학', '학년', '총점'], inplace=True)
score
```

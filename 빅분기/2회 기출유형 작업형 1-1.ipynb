## 문제1

* 데이터셋(basic1.csv)의 'f5' 컬럼을 기준으로 상위 10개의 데이터를 구하고,
* 'f5'컬럼 10개 중 최소값으로 데이터를 대체한 후,
* 'age'컬럼에서 80 이상인 데이터의'f5 컬럼 평균값 구하기

```python
import pandas as pd

df = pd.read_csv('../dataset/basic1.csv')
df.head()
```

## Solution

```python
# f5컬럼을 기준으로 내림차순 정렬
df = df.sort_values('f5', ascending=False)
df.head(10)
```

```python
# 최소값 찾기
min_val = df['f5'][:10].min()
min_val
```

```python
df['f5'][:10] = min_val
df.head(10)
```

```python
# 80세 이상의 f5컬럼 평균
print(df[df['age']>=80]['f5'].mean())
```

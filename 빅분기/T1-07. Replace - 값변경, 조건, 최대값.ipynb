##### 라이브러리 및 데이터 불러오기

```python
import pandas as pd

df = pd.read_csv('../dataset/basic1.csv')
df.head()
```

##### ESFJ 값을 가진 데이터 확인

```python
df[df['f4']=='ENFJ']
```

##### 값 변경하기

```python
df['f4'] = df['f4'].replace('ENFJ', 'ISFJ')
df.head()
```

##### 2개의 조건에 맞는 값중 age컬럼의 최대값

```python
# 'city'가 '경기'이면서 'f4'가 'ISFJ'인 데이터 중 'age'컬럼의 최대값을 출력
df = df[(df['city'] == '경기')&(df['f4'] == 'ISFJ')]
df
```
